## 🧹 Preprocessing Pipeline

We built a reusable preprocessing function that we applied consistently across all submissions.

```python
def encode_data(train_df, test_df):
    """Label encode all categorical features consistently across train and test."""
    train_enc = train_df.copy()
    test_enc  = test_df.copy()

    cat_cols = train_enc.select_dtypes(include='object').columns.tolist()
    # Remove identifier and target columns
    for col in ['uniqueid', 'bank_account']:
        if col in cat_cols:
            cat_cols.remove(col)

    le = LabelEncoder()
    for col in cat_cols:
        combined = pd.concat([train_enc[col], test_enc[col]], axis=0)
        le.fit(combined)
        train_enc[col] = le.transform(train_enc[col])
        test_enc[col]  = le.transform(test_enc[col])

    return train_enc, test_enc

print('✅ Encoding function defined.')
```

```python
def make_submission(preds, test_df, filename):
    """Create and download a Zindi-formatted submission CSV."""
    sub = pd.DataFrame({
        'uniqueid'     : test_df['uniqueid'] + ' x ' + test_df['country'],
        'bank_account' : preds
    })
    sub.to_csv(filename, index=False)
    files.download(filename)
    print(f'📥 Downloaded: {filename}  |  Shape: {sub.shape}')
    print(sub['bank_account'].value_counts())
    return sub

print('✅ Submission function defined.')
```

# 🔧 Feature Engineering Pipeline

After submission #07, we committed to serious feature engineering. This became the most impactful part of our journey.

```python
def feature_engineering_v1(df):
    """Version 1 — Basic feature engineering."""
    df = df.copy()

    # Age bins
    df['age_bin'] = pd.cut(df['age_of_respondent'],
                            bins=[0, 25, 35, 45, 55, 65, 100],
                            labels=[0, 1, 2, 3, 4, 5]).astype(int)

    # Household size buckets
    df['hh_size_bin'] = pd.cut(df['household_size'],
                                bins=[0, 2, 4, 7, 100],
                                labels=[0, 1, 2, 3]).astype(int)

    # Interaction: urban + cellphone
    df['urban_cell'] = ((df['location_type'] == 'Urban') &
                         (df['cellphone_access'] == 'Yes')).astype(int)

    # Interaction: education + gender
    df['edu_gender'] = df['education_level'].astype(str) + '_' + df['gender_of_respondent'].astype(str)

    return df


def feature_engineering_v2(df):
    """Version 2 — Full feature engineering pipeline."""
    df = feature_engineering_v1(df)

    # Finer age groups
    df['age_fine_bin'] = pd.cut(df['age_of_respondent'],
                                 bins=[0,18,25,30,35,40,50,60,100],
                                 labels=range(8)).astype(int)

    # Is head of household
    df['is_head'] = (df['relationship_with_head'] == 'Head of Household').astype(int)

    # Is employed (any formal job)
    employed_types = ['Government Dependent', 'Formally employed Government',
                      'Formally employed Private', 'Self employed']
    df['is_employed'] = df['job_type'].isin(employed_types).astype(int)

    # Has formal education
    formal_edu = ['Secondary education', 'Tertiary education',
                  'Vocational/Specialised training']
    df['has_formal_edu'] = df['education_level'].isin(formal_edu).astype(int)

    # Job type grouping (reduced categories)
    job_group_map = {
        'Farming and Fishing': 'informal',
        'Self employed': 'self_employed',
        'Formally employed Private': 'formal',
        'Formally employed Government': 'formal',
        'Government Dependent': 'dependent',
        'Remittance Dependent': 'dependent',
        'Other Income': 'other',
        'No Income': 'no_income',
        'Dont Know/Refuse to answer': 'unknown'
    }
    df['job_group'] = df['job_type'].map(job_group_map).fillna('other')

    # Marital simplification
    df['is_married'] = df['marital_status'].str.contains('Married|Living', na=False).astype(int)

    return df


print('✅ Feature engineering functions defined.')
```

```python
def add_group_stats(train_df, test_df, group_col, target='bank_account_num'):
    """Add country-level and group-level aggregation features."""
    stats = train_df.groupby(group_col)[target].agg(['mean', 'count', 'std']).reset_index()
    stats.columns = [group_col, f'{group_col}_bank_rate',
                     f'{group_col}_count', f'{group_col}_bank_std']
    train_out = train_df.merge(stats, on=group_col, how='left')
    test_out  = test_df.merge(stats, on=group_col, how='left')
    return train_out, test_out

print('✅ Group stats function defined.')
```

# 📝 Submission #10 — Undersampling

**Strategy:** Try undersampling the majority class instead of oversampling minority.

# 📝 Submission #18 — Stratified K-Fold CV Ensemble

**Strategy:** Run the ensemble inside a full Stratified K-Fold for better OOF predictions and test averaging.

**Zindi Score:** ~0.9020 ⬆️

```python
# ─── Submission #18 ───────────────────────────────────────────────────────────
def run_cv_ensemble(X, y, X_test, n_splits=5):
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
    oof    = np.zeros(len(X))
    test_p = np.zeros(len(X_test))
    cv_scores = []

    for fold, (tr_idx, val_idx) in enumerate(skf.split(X, y)):
        X_tr_f, X_val_f = X.iloc[tr_idx], X.iloc[val_idx]
        y_tr_f, y_val_f = y.iloc[tr_idx], y.iloc[val_idx]

        lgbm = LGBMClassifier(**best_lgbm_params, verbose=-1)
        xgb  = XGBClassifier(n_estimators=500, use_label_encoder=False,
                               eval_metric='logloss', random_state=42, verbosity=0)

        lgbm.fit(X_tr_f, y_tr_f)
        xgb.fit(X_tr_f, y_tr_f)

        val_proba = (lgbm.predict_proba(X_val_f)[:,1] + xgb.predict_proba(X_val_f)[:,1]) / 2
        oof[val_idx] = val_proba

        test_proba = (lgbm.predict_proba(X_test)[:,1] + xgb.predict_proba(X_test)[:,1]) / 2
        test_p += test_proba / n_splits

        fold_score = accuracy_score(y_val_f, (val_proba > 0.5).astype(int))
        cv_scores.append(fold_score)
        print(f'  Fold {fold+1}: {fold_score:.4f}')

    print(f'\nOOF Accuracy : {accuracy_score(y, (oof > 0.5).astype(int)):.4f}')
    print(f'CV Mean      : {np.mean(cv_scores):.4f} ± {np.std(cv_scores):.4f}')
    return (test_p > 0.5).astype(int)

test_preds_18 = run_cv_ensemble(X_full, y, X_test_full)
sub_18 = make_submission(test_preds_18, test, 'submission_18_stratified_cv_ensemble.csv')
```

# 📝 Submissions #19-#21 — Target Encoding & Stacking

**#19 Strategy:** Replace label encoding with target encoding for high-cardinality categoricals.

**#20 Strategy:** Stack LGBM + XGB + RF with a Logistic Regression meta-learner.

**#21 Strategy:** Add CatBoost to the stack as a 4th base model.

**Scores:** #19 ~0.9015 | #20 ~0.9022 ⬆️ | #21 ~0.9025 ⬆️

```python
# ─── Submission #19 — Target Encoding ────────────────────────────────────────
def target_encode(train_df, test_df, col, target, smoothing=10):
    """Target encode a column with smoothing to prevent overfitting."""
    global_mean = train_df[target].mean()
    agg = train_df.groupby(col)[target].agg(['mean', 'count'])
    smooth = (agg['count'] * agg['mean'] + smoothing * global_mean) / (agg['count'] + smoothing)
    train_df[f'{col}_te'] = train_df[col].map(smooth).fillna(global_mean)
    test_df[f'{col}_te']  = test_df[col].map(smooth).fillna(global_mean)
    return train_df, test_df

train_te = train_fe2_enc.copy()
test_te  = test_fe2_enc.copy()
train_te['bank_account_num'] = y.values

for col in ['country', 'job_type', 'education_level', 'marital_status', 'relationship_with_head']:
    if col in train_te.columns:
        train_te, test_te = target_encode(train_te, test_te, col, 'bank_account_num')

features_19 = [c for c in train_te.columns
               if c not in ['uniqueid', 'bank_account', 'bank_account_num', 'year']]
X_19 = train_te[features_19]
X_test_19 = test_te[features_19]

X_tr19, X_val19, y_tr19, y_val19 = train_test_split(X_19, y, test_size=0.2, random_state=42, stratify=y)
model_19 = LGBMClassifier(**best_lgbm_params, verbose=-1)
model_19.fit(X_tr19, y_tr19)
print(f'Submission #19 | Validation Accuracy: {accuracy_score(y_val19, model_19.predict(X_val19)):.4f}')
sub_19 = make_submission(model_19.predict(X_test_19), test, 'submission_19_target_encoding.csv')
```

```python
# ─── Submission #20 — Stacking ────────────────────────────────────────────────
base_estimators_20 = [
    ('lgbm', LGBMClassifier(**best_lgbm_params, verbose=-1)),
    ('xgb',  XGBClassifier(n_estimators=500, use_label_encoder=False,
                            eval_metric='logloss', random_state=42, verbosity=0)),
    ('rf',   RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1))
]

stacking_20 = StackingClassifier(
    estimators=base_estimators_20,
    final_estimator=LogisticRegression(max_iter=1000),
    cv=5, passthrough=True, n_jobs=-1
)

stacking_20.fit(X_tr17, y_tr17)
print(f'Submission #20 | Validation Accuracy: {accuracy_score(y_val17, stacking_20.predict(X_val17)):.4f}')
sub_20 = make_submission(stacking_20.predict(X_test_full), test, 'submission_20_stacking_3models.csv')
```

```python
# ─── Submission #27 — FINAL CHAMPION ENSEMBLE ─────────────────────────────────
print('=' * 60)
print('  SUBMISSION #27 — FINAL BEST ENSEMBLE')
print('=' * 60)

skf_final = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

oof_lgbm = np.zeros(len(X_24))
oof_xgb  = np.zeros(len(X_24))
oof_cb   = np.zeros(len(X_24))

test_lgbm = np.zeros(len(X_test_24))
test_xgb  = np.zeros(len(X_test_24))
test_cb   = np.zeros(len(X_test_24))

for fold, (tr_idx, val_idx) in enumerate(skf_final.split(X_24, y)):
    X_tr_f, X_val_f = X_24.iloc[tr_idx], X_24.iloc[val_idx]
    y_tr_f, y_val_f = y.iloc[tr_idx], y.iloc[val_idx]

    # LGBM
    m_lgbm = LGBMClassifier(**final_best_params, verbose=-1)
    m_lgbm.fit(X_tr_f, y_tr_f)
    oof_lgbm[val_idx]  = m_lgbm.predict_proba(X_val_f)[:, 1]
    test_lgbm         += m_lgbm.predict_proba(X_test_24)[:, 1] / 5

    # XGBoost
    m_xgb = XGBClassifier(n_estimators=700, use_label_encoder=False,
                           eval_metric='logloss', random_state=42,
                           learning_rate=0.05, max_depth=6, verbosity=0)
    m_xgb.fit(X_tr_f, y_tr_f)
    oof_xgb[val_idx]  = m_xgb.predict_proba(X_val_f)[:, 1]
    test_xgb         += m_xgb.predict_proba(X_test_24)[:, 1] / 5

    # CatBoost
    m_cb = CatBoostClassifier(iterations=500, learning_rate=0.05,
                               depth=6, random_seed=42, verbose=0)
    m_cb.fit(X_tr_f, y_tr_f)
    oof_cb[val_idx]  = m_cb.predict_proba(X_val_f)[:, 1]
    test_cb         += m_cb.predict_proba(X_test_24)[:, 1] / 5

    lgbm_f = accuracy_score(y_val_f, (oof_lgbm[val_idx] > 0.5).astype(int))
    xgb_f  = accuracy_score(y_val_f, (oof_xgb[val_idx] > 0.5).astype(int))
    cb_f   = accuracy_score(y_val_f, (oof_cb[val_idx] > 0.5).astype(int))
    print(f'  Fold {fold+1} — LGBM: {lgbm_f:.4f} | XGB: {xgb_f:.4f} | CB: {cb_f:.4f}')

# Weighted ensemble
w_lgbm, w_xgb, w_cb = 0.50, 0.30, 0.20

oof_ensemble  = w_lgbm*oof_lgbm + w_xgb*oof_xgb + w_cb*oof_cb
test_ensemble = w_lgbm*test_lgbm + w_xgb*test_xgb + w_cb*test_cb

oof_final_preds  = (oof_ensemble  > 0.5).astype(int)
test_final_preds = (test_ensemble > 0.5).astype(int)

print(f'\n{"="*60}')
print(f'  OOF LGBM     : {accuracy_score(y, (oof_lgbm  > 0.5).astype(int)):.4f}')
print(f'  OOF XGBoost  : {accuracy_score(y, (oof_xgb   > 0.5).astype(int)):.4f}')
print(f'  OOF CatBoost : {accuracy_score(y, (oof_cb    > 0.5).astype(int)):.4f}')
print(f'  OOF ENSEMBLE : {accuracy_score(y, oof_final_preds):.4f}  ← BEST')
print(f'{"="*60}')

sub_27 = make_submission(test_final_preds, test, 'submission_27_FINAL_BEST.csv')
print('\n🏆 FINAL BEST SUBMISSION CREATED!')
```

## 🎓 Key Lessons Learned

After 27 submissions and many late nights, here's what actually moved the needle:

| Technique | Impact | Notes |
|-----------|--------|-------|
| Proper label encoding (combined fit) | ✅ Small | Subtle bug fix with real impact |
| Dropping `year` | ✅ Small | Metadata, not a signal |
| Feature engineering (age bins, job groups) | ✅✅ Medium | Most impactful single change |
| Group statistics (country bank rate) | ✅✅ Medium | Excellent domain-informed feature |
| SMOTE / Undersampling | ❌ Negative | Hurt accuracy; resampling ≠ always good |
| Optuna hyperparameter tuning | ✅ Medium | First time breaking 0.90 |
| Ensemble (LGBM + XGB + CatBoost) | ✅✅ Medium | Diversity of models helps |
| Stratified K-Fold CV | ✅ Small | Better validation, more reliable OOF |
| Test-Time Augmentation | ❌ Slight negative | Not designed for tabular data |
| Target encoding | ✅ Small | Marginal improvement |

### 🔑 Golden Rules We Discovered:

1. **Feature engineering beats hyperparameter tuning** — always do FE first.
2. **OOF accuracy is your best guide** — trust it over a single val split.
3. **Resampling for accuracy-scored competitions often backfires** — the metric rewards majority class.
4. **Diverse ensembles beat any single model** — LGBM + XGB + CatBoost have different strengths.
5. **Domain knowledge matters** — grouping job types and understanding the African financial context improved features significantly.
6. **More submissions = more learning** — every failure narrowed the search space.
